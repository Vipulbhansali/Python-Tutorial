```python
import requests
from bs4 import BeautifulSoup

# Make a request
response = requests.get('https://github.com/topics/3d')

# Create a soup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find all the repository tags in the document
repo_tags = soup.find_all('h3', class_='f3 color-fg-muted text-normal lh-condensed')

# Create lists to store the information we want to extract
names = []
repo_names = []
star_counts = []

# Iterate over the repository tags and extract the information we want
for repo in repo_tags:
    # Extract the username and repository name
    a_tags = repo.find_all('a')
    username = a_tags[0].text.strip()
    repo_name = a_tags[1].text.strip()

    # Find the star count element and extract the text from it
    star_count_tag = soup.find('a', href=f'/{username}/{repo_name}/stargazers')
    if star_count_tag is None:
        star_count = '0'
    else:
        star_count = star_count_tag.find('span', class_='Counter js-social-count').text.strip()

    # Add the information to the lists
    names.append(username)
    repo_names.append(repo_name)
    star_counts.append(star_count)

# Print the information we extracted
for i in range(len(names)):
    print(f'{names[i]}/{repo_names[i]}: {star_counts[i]}')
```

```python
import requests
from bs4 import BeautifulSoup

# Make a request
response = requests.get('https://github.com/topics/3d')

# Create a soup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find all the repository tags in the document
repo_tags = soup.find_all('h3', class_='f3 color-fg-muted text-normal lh-condensed')

# Loop over each repository tag and extract the relevant information
for repo in repo_tags:
    # Find the username and repository name elements and extract the text from them
    repo_elem = repo.find('a', class_='text-bold')
    full_name = repo_elem.text.strip().split('/')
    username = full_name[0]
    repo_name = full_name[1]

    # Find the star count element and extract the text from it
    header = repo.find('div', class_='d-inline-block float-sm-right')
    star_tag = header.find_all('span', {'class': 'Counter js-social-count'})
    star_count = star_tag[0].text.strip()

    # Print the extracted information
    print(f'{username}/{repo_name}: {star_count}')
```

```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm

# Make a request
response = requests.get('https://github.com/topics/3d')

# Create a soup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find all the repository tags in the document
repo_tags = soup.find_all('h3', class_='f3 color-text-secondary text-normal lh-condensed')

# Loop through each repository tag and extract the username, repo name, and star count
for repo in tqdm(repo_tags):
    username, repo_name = repo.find('a').get('href').split('/')[1:3]
    repo_url = f'https://github.com/{username}/{repo_name}'
    star_count = 0
    try:
        response = requests.get(repo_url)
        soup = BeautifulSoup(response.text, 'html.parser')
        star_tag = soup.find('a', href=f'/{username}/{repo_name}/stargazers')
        if star_tag:
            star_count = star_tag.text.strip()
            if star_count[-1] == 'k':
                star_count = float(star_count[:-1]) * 1000
            else:
                star_count = int(star_count)
    except:
        pass
    print(f'{username}/{repo_name}: {star_count} stars')
```
